```python
import random
```

```python
#random.random()
#Returns a random float between 0.0 (inclusive) and 1.0 (exclusive).
random_1 = random.random() 
print(random_1)

#random.randint(a, b)
#Returns a random integer N such that a < N > b (both endpoints are inclusive).
random_2 = random.randint(1, 10) 
print(random_2)
                                                        
#random.uniform(a, b)
#Returns a random float N such that a < N > b. (More flexible than random() for specifying ranges).
random_3 = random.uniform(10.5, 20.5) 
print(random_3)
    
#random.randrange(start, stop, step)
#Returns a random integer from a specified range (similar to the range() function). The stop value is exclusive.
random_4 = random.randrange(0, 10, 2)
print(random_4)
```

**Day 4 Project: Rock Paper Scissors**

[World Rock Paper Scissors Association](https://wrpsa.com/)

```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''
paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''
scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

user_choice = (input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors\n0/1/2:"))
computers_choice = random.randint(0,2)

if not user_choice.isdigit():
  print("You've entered an invalid value, try again and choose a number between 0-2.")
else:
  choice = int(user_choice)
  if choice > 2:
    print("You've entered an invalid number, try again and choose a number between 0-2.")
  elif user_choice == 0 or choice == 1 or choice == 2:
    if user_choice == 0:
      print(f"You chose: Rock {rock}")
    elif user_choice == 1:
      print(f"You chose: Paper {paper}")
    elif user_choice == 2:
      print(f"You chose: Scissors {scissors}")

    if computers_choice == 0:
      print(f"The computer chose: Rock {rock}")
      if user_choice == 2:
        print("You lose, Rock wins against scissors.")
      elif user_choice == 0:
        print("It's a draw!")
      else:
        print("You win!")
    elif computers_choice == 1:
      print(f"The computer chose: Paper {paper}")
      if user_choice == 0:
        print("You lose, Paper wins against rock.")
      elif user_choice == 0:
        print("It's a draw!")
      else:
        print("You win!")

    elif computers_choice == 2:
      print(f"The computer chose: Scissors {scissors}")
      if user_choice == 1:
        print("You lose, Scissors win against paper.")
      elif user_choice == 2:
        print("It's a draw!")
      else:
        print("You win!")          
```
