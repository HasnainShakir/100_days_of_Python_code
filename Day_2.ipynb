**String (str):** Used for textual data, consisting of characters enclosed in single or double quotes (e.g., "Hello").
**Integer (int): whole number** Used for whole numbers (positive, negative, or zero) without any decimal point (e.g., 42).
**Float (float):** Used for numbers with a decimal point or for scientific notation (e.g., 3.14 or 1.0).
**Boolean (bool):** Used for the two logical states: True or False, essential for decision-making and logic flow.

**Mathematical Operations in Python**

```python
print(12 + 3)            #Addition
print(12 - 3)            #Subtraction
print(12 * 3)            #Multiplication
print(12 / 5)            #Division (float)
print(12 // 5)           #Division, this will remove all floats and the type will be integer
print(type(12 / 5))      #Division (float)
print(type(12 // 5))     #Division, this will remove all floats and the type will be integer
print(2 ** 5)            #exponent 
# Python follows PEMDAS, and calculations are done from left to right. it means first it will process the left one if equally important
```

**Day 2 Project: Tip Calculator**

```python
print("Welcome to the tip calculator!")
total_bill = float(input("What was the total bill? $"))

tip = int(input("How much tip would you like to give? 10, 12, or 15?"))
split = int(input("How many people to split the bill?"))
total = ("{:.2f}".format((((total_bill * (tip / 100)) + total_bill) / split)))
print(f"Each person should pay: ${total}")
```
